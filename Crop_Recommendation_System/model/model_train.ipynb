```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Activation
from keras.optimizers import SGD
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
```

```python
df = pd.read_csv("cropdata.csv")
```

```python
le = preprocessing.LabelEncoder()
l1 = df["Soil"]
le.fit(l1)
newsoil = le.transform(l1)
df["Soil"]=newsoil

l2 = df["Month"]
le.fit(l2)
df["Month"]=le.transform(l2)

l3 = df["State"]
le.fit(l3)
df["State"]=le.transform(l3)
```

```python
df['Month']
```

```python
df.head()
```

```python
df = pd.DataFrame(data=df.iloc[:, :].values, columns=["Soil", "Month", "State",
                                                       "Rice", "Wheat", "Cotton", "Sugarcane", "Tea", "Ragi", "Maize", "Millet", "Barley"])
```

```python
df[:50]
```

```python
feat = pd.DataFrame(
    {"Soil": df["Soil"], "Month": df["Month"], "State": df["State"]})
labels = pd.DataFrame(data=df.iloc[:, 3:], columns=["Rice", "Wheat", "Cotton", "Sugarcane", "Tea", "Ragi", "Maize", "Millet", "Barley"])
```

```python
feat.head()
```

```python
labels.head()
```

```python
(trainData, testData, trainLabels, testLabels) = train_test_split(
    feat, labels, test_size=0.25, random_state=42)
print(trainData.values)
```

```python
print(trainData.shape, testData.shape, trainLabels.shape, testLabels.shape)
```

```python
nn = Sequential()
nn.add(Dense(32, activation="relu", input_shape=(3,)))
nn.add(Dense(24, activation="sigmoid"))
nn.add(Dense(9, activation="sigmoid"))
print(nn.summary())
```

```python
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
nn.fit(trainData.values, trainLabels.values,epochs=50, batch_size=10, verbose=1)
```

```python
(loss, accuracy) = nn.evaluate(testData.values,
                                  testLabels.values,	batch_size=40, verbose=1)

print("[INFO] loss={:.4f}, accuracy: {:.4f}%".format(loss, accuracy * 100))
```
